```python
import pandas as pd
df = pd.read_csv("./Ejercicio_3/AB_NYC_2019.csv", index_col= ["id"])
df
df.head(5)
```

Problematicas


```python
# Rellenar reviews_per_month con 0 donde number_of_reviews es 0
df['reviews_per_month'] = df.apply(
    lambda row: 0 if pd.isnull(row['reviews_per_month']) and row['number_of_reviews'] == 0 else row['reviews_per_month'],
    axis=1
)

# Rellenar las columnas name y host_name con 'Desconocido' (o puedes eliminarlas si prefieres)
df['name'].fillna('Desconocido', inplace=True)
df['host_name'].fillna('Desconocido', inplace=True)

# Revisar si quedan nulos
print(df.isnull().sum())
```

```python
print(df.isnull().sum())
```

```python
# Revisar valores extremos en price
print(df['price'].describe())

# Eliminar registros con precios muy altos (ej. más de 1000 USD por noche)
df = df[df['price'] <= 1000]

# Filtrar minimum_nights absurdos (ej. más de 365 noches)
df = df[df['minimum_nights'] <= 365]
```

```python
# Convertir last_review a formato datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# Limpiar espacios y pasar a minúscula
df['room_type'] = df['room_type'].str.strip().str.lower()
df['neighbourhood_group'] = df['neighbourhood_group'].str.strip().str.lower()
df['neighbourhood'] = df['neighbourhood'].str.strip().str.lower()
```

```python
df.to_csv("AB_NYC_2019_limpio.csv", index=False)
```
